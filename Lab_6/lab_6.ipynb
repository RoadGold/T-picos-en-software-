```python
from datetime import date

class Cliente:
    def __init__(self, dni, nombre, sexo, fecha_nacimiento):
        self.dni = dni
        self.nombre = nombre
        self.sexo = sexo
        self.fecha_nacimiento = fecha_nacimiento

    def mostrar(self):
        print(f"DNI: {self.dni}, Nombre: {self.nombre}, Sexo: {self.sexo}, Fecha de Nacimiento: {self.fecha_nacimiento}")

    def calcular_edad(self):
        today = date.today()
        edad = today.year - self.fecha_nacimiento.year - ((today.month, today.day) < (self.fecha_nacimiento.month, self.fecha_nacimiento.day))
        return edad


class CuentaBancaria:
    def __init__(self, numero_cuenta, saldo, tipo_cuenta):
        self.numero_cuenta = numero_cuenta
        self.saldo = saldo
        self.tipo_cuenta = tipo_cuenta

    def depositar(self, cantidad):
        self.saldo += cantidad

    def retirar(self, cantidad):
        if cantidad <= self.saldo:
            self.saldo -= cantidad
        else:
            print("Saldo insuficiente")

    def consultar_saldo(self):
        return self.saldo

    def obtener_tipo_cuenta(self):
        return self.tipo_cuenta
```

```python
def main():
    # Crear un cliente
    cliente1 = Cliente(dni="12345678", nombre="Juan Perez", sexo="Masculino", fecha_nacimiento=date(1990, 1, 1))
    cliente1.mostrar()
    print(f"Edad: {cliente1.calcular_edad()} años")

    # Crear una cuenta bancaria
    cuenta1 = CuentaBancaria(numero_cuenta="0001", saldo=1000.0, tipo_cuenta="Ahorros")
    print(f"Saldo inicial: {cuenta1.consultar_saldo()}")

    # Realizar operaciones
    cuenta1.depositar(500.0)
    print(f"Saldo después del depósito: {cuenta1.consultar_saldo()}")

    cuenta1.retirar(300.0)
    print(f"Saldo después del retiro: {cuenta1.consultar_saldo()}")

if __name__ == "__main__":
    main()
```

```python
import sqlite3

def crear_bd():
    conn = sqlite3.connect('banco.db')
    cursor = conn.cursor()

    # Crear tabla Cliente
    cursor.execute('''CREATE TABLE Cliente (
                        dni TEXT PRIMARY KEY,
                        nombre TEXT NOT NULL,
                        sexo TEXT NOT NULL,
                        fecha_nacimiento DATE NOT NULL
                    )''')

    # Crear tabla CuentaBancaria
    cursor.execute('''CREATE TABLE CuentaBancaria (
                        numero_cuenta TEXT PRIMARY KEY,
                        saldo REAL NOT NULL,
                        tipo_cuenta TEXT NOT NULL,
                        dni_cliente TEXT,
                        FOREIGN KEY (dni_cliente) REFERENCES Cliente (dni)
                    )''')

    conn.commit()
    conn.close()

crear_bd()
```
